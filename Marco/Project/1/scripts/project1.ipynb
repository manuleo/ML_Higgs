```python
def compute_stoch_gradient(y, tx, w):
    """Compute a stochastic gradient from just few examples n and their corresponding y_n labels."""
    e = y - np.dot(tx,w)
    return -1/(2*len(y))*np.dot(tx.transpose(),e)


def stochastic_gradient_descent(y, tx, initial_w, batch_size, max_iters, gamma):
    """Stochastic gradient descent algorithm."""
    # Define parameters to store w and loss
    #ws = [initial_w]
    loss = np.inf
    w = initial_w
    for n_iter in range(max_iters):
        for minibatch_y, minibatch_tx in batch_iter(y, tx, batch_size):
            y_i = minibatch_y
            tx_i = minibatch_tx
        G = compute_stoch_gradient(y_i, tx_i,w)
        new_w = w-gamma*G
        new_loss = compute_loss(y, tx,new_w)
        new_w = w-gamma*G
        #ws.append(w)
        if new_loss <= loss:
            loss , w = new_loss , new_w
            gamma *=1.8
            print("SGD({bi}/{ti}): loss={l} ; gamma={g}; gamma aumenta".format(
              bi=n_iter, ti=max_iters - 1, l=loss, g= gamma))
        else:
            gamma *=0.3
            print("SGD({bi}/{ti}): loss={l} ; gamma={g}; gamma diminuisce".format(
                  bi=n_iter, ti=max_iters - 1, l=loss, g= gamma))
    return
```
